```python
import pandas as pd
```

```python
df = pd.read_csv("../raw_data/laps.csv")
df
```

```python
df["TotalLaps"] = 0
```

```python
df
```

```python
coucou = df.groupby(by = ["Year", "Location"], as_index=False)["LapNumber"].max().rename(columns={"LapNumber":"TotalLaps"})
coucou["Year_Location"] = coucou["Year"].map(str) + coucou["Location"]
coucou
```

```python
df["Year_Location"] = df["Year"].map(str) + df["Location"]
```

```python
df
```

```python
final_df = df.merge(coucou, on="Year_Location")
```

```python
final_df.head()
```

```python
final_df.drop(columns=["TotalLaps_x", "Year_Location", "Year_y", "Location_y"], inplace=True)
```

```python
final_df["RaceProgress"] = final_df["LapNumber"] / final_df["TotalLaps_y"]
```

```python
final_df
```

```python
real_final_df = final_df.rename(columns={"Location_x": "Location",
                         "Year_x": "Year",
                         "TotalLaps_y": "TotalLaps"})
```

```python
real_final_df
```

```python
pit_stops = real_final_df[real_final_df["PitInTime"].notna()]
pit_stops
```

```python
early_pit_stops = pit_stops[pit_stops["RaceProgress"] < 0.1]
```

```python
early_pit_stops["Driver_Location_Year"] = early_pit_stops["Driver"] + early_pit_stops["Location"] + early_pit_stops["Year"].map(str)
```
