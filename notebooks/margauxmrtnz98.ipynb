# Création DataSet Laps

```python
#2018
df_laps2018 = pd.DataFrame()
season = fastf1.get_event_schedule(2018)
locations = season['Location'].values.tolist()

for location in locations :
    session = fastf1.get_session(2018, location, 'R')
    session.load()
    session.laps['Location'] = location
    session.laps['Year'] = 2018
    df_laps2018 = pd.concat([df_laps2018,session.laps],ignore_index=True)
df_laps2018
```
