# Création Dataset Result

```python
years = [2017,2018,2019,2020,2021,2022]
```

```python
#Résultats finaux par saisons
df_results_per_year = pd.DataFrame()

for year in years : 
    season = fastf1.get_event_schedule(year)
    locations = season['Location'].values.tolist()
    df_results = pd.DataFrame()

    for location in locations :
        session = fastf1.get_session(year, location, 'R')
        session.load()
        points_per_team = session.results[['TeamId','Points']].groupby('TeamId').sum('Points').reset_index()
        df_results = pd.concat([df_results,points_per_team],ignore_index=True).groupby('TeamId').sum('Points').sort_values('Points',ascending=False).reset_index()
        df_results['Year']= year
        df_results['Ranking'] = df_results['Points'].rank(ascending=False)
    
    df_results_per_year = pd.concat([df_results_per_year,df_results],ignore_index=True)
    
df_results_per_year
```

```python
df_results_per_year.to_csv('results.csv')
```

# Trouver la position de l'écurie au précédent Grand Prix

```python
laps_copy = pd.read_csv('../raw_data/laps.csv').copy()
laps_copy.head()
```

```python
for index, row in laps_copy.iterrows():
    
    #Infos Grand Prix
    year = row['Year']
    location = row['Location']
    team = row['Team']
    
    #Infos Saison
    season = fastf1.get_event_schedule(year)
    locations = season['Location'].values.tolist()
    
    results = pd.read_csv('results.csv')
    
    #Si c'est le 1er Grand Prix de l'année - on récupère le classement final de la saison précédente :
    if location == locations[0] :
        session = fastf1.get_session(year-1, location, 'R')
        session.load()
        results = session.results[['TeamName','Points']].groupby('TeamName').sum('Points').sort_values('Points',ascending=False).reset_index()
        results['Ranking'] = results['Points'].rank(ascending=False)
        laps_copy['Last_TeamRanking'] = results['Ranking'][(results['TeamName'] == team) & (results['Year'] == year)]
        
    #Sinon, classement après le dernier Grand Prix
    else :
        for i in locations[0:locations.index(location)] :
            session = fastf1.get_session(year, i, 'R')
            session.load()
        points_per_team = session.results[['TeamId','Points']].groupby('TeamId').sum('Points').reset_index()
        df_results = pd.concat([df_results,points_per_team],ignore_index=True).groupby('TeamId').sum('Points').sort_values('Points',ascending=False).reset_index()
        df_results['Year']= year
        df_results['Ranking'] = df_results['Points'].rank(ascending=False)
    
            
        
        
```
