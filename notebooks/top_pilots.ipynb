```python
import pandas as pd
```

```python
df = pd.read_csv('../raw_data/laps.csv')
top_10_df = df[~(df['Position'] >= 10)]
top_10_df
```

```python
df['LocationYear'] = df['Location'] + ' ' + df['Year'].astype(str)
lap_total = df.groupby('LocationYear')['LapNumber'].max()
lap_total
```

```python
result = df[['LapNumber']].isin(lap_total)
result
```

```python
result = df[df['LapNumber'].isin(lap_total)].loc[df['Position'].between(1, 10)]
result
```

```python
top_laps_df = df[df['LapNumber'] == lap_total]
top_laps_df
```

```python
top_10_positions = df[~(df['Position'] >= 10)]
top_10_positions
```
