```python
import pandas as pd

df = pd.read_csv('../raw_data/laps.csv')

df['LocationYear'] = df['Location'] + ' ' + df['Year'].astype(str)

wet_gp = df[(df['Compound'] == 'INTERMEDIATE') | (df['Compound'] == 'WET')]['LocationYear'].unique()

rainy_gp = df['LocationYear'].isin(wet_gp)
df = df[~rainy_gp]
df
```
