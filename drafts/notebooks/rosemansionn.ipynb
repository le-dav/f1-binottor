```python
years = [2018]
weather_df_2018 = pd.DataFrame()

season = fastf1.get_event_schedule(2018)
locations = season['Location'].values.tolist()

for location in locations :
    session = fastf1.get_session(2018, location, 'R')
    session.load()
    session.weather_data['Location'] = location
    session.weather_data['Year'] = 2018
    weather_df_2018 = pd.concat([weather_df_2018, session.weather_data], ignore_index=True)
weather_df_2018
```
